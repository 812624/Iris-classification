# Importing all required dependencies

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Getting the data

```python
# Renaming columns for simplicity
data = pd.read_csv("/home/sakshi/Desktop/Iris1.csv", names = ['Sl', 'Sw', 'Pl', 'Pw', 'Species'])
```

```python
data
```

```python
# mapping data into arrays
s = np.asarray([1,0,0])
ve = np.asarray([0,1,0])
vi = np.asarray([0,0,1])
data['Species'] = data['Species'].map({'Iris-setosa' : s, 'Iris-versicolor' : ve, 'Iris-virginica' : vi})
```

```python
data
```

# Shuffling the data

```python
import random
data=data.loc[np.random.permutation(len(data))]
```

```python
data
```

```python
# Setting indices properly
data = data.reset_index(drop = True)
```

# Dividing dataset into training and testing data

```python
# Training data
x_input=data.ix[0:105,['Sl', 'Sw', 'Pl', 'Pw']]
temp=data['Species']
y_input=temp[0:106]
# Test data
x_test=data.ix[106:149,['Sl', 'Sw', 'Pl', 'Pw']]
y_test=temp[106:150]
```

# Getting the right placeholders declared

```python
# Placeholders and variables. input has 4 features and output has 3 classes
x=tf.placeholder(tf.float32,shape=[None,4])
y_=tf.placeholder(tf.float32,shape=[None, 3])
# Weight and bias
W=tf.Variable(tf.zeros([4,3]))
b=tf.Variable(tf.zeros([3]))
```

# Model

```python
# Softmax function for multiclass classification
y = tf.nn.softmax(tf.matmul(x, W) + b)
```

```python
# Loss function
cross_entropy = tf.reduce_mean(-tf.reduce_sum(y_ * tf.log(y), reduction_indices=[1]))
```

```python
# Optimizer
train_step = tf.train.AdamOptimizer(0.01).minimize(cross_entropy)
```

# Calculating Accuracy

```python
# Calculating accuracy of our model 
correct_prediction = tf.equal(tf.argmax(y,1), tf.argmax(y_,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

# Initialising the session parameters

```python
# Session parameters
sess = tf.InteractiveSession()
# Initialising variables
init = tf.global_variables_initializer()
sess.run(init)
# Number of interations
epoch=2000
```

```python
for step in xrange(epoch):
   _, c=sess.run([train_step,cross_entropy], feed_dict={x: x_input, y_:[t for t in y_input.as_matrix()]})
   if step % 500==0 :
        print c
```

# Testing on own data

```python
a = data.loc[130,['Sl', 'Sw', 'Pl', 'Pw']]
b = a.values.reshape(1,4)
largest = sess.run(tf.arg_max(y,1), feed_dict={x: b})[0]
if largest==0:
    print "flower is :Iris-setosa"
elif largest==1:
    print "flower is :Iris-versicolor"
else :
    print "flower is :Iris-virginica"
```
